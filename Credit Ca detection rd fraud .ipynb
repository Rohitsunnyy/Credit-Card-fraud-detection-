```python
#as you can see from the data that Time and Amount has to be scaled as the remaning columns are already scaled 
#using Robustscaler to scale the data 
from sklearn.preprocessing import RobustScaler

scaler =RobustScaler()

df['Scaled_Time'] = scaler.fit_transform(df['Time'].values.reshape(-1,1))

df['Scaled_Amount'] = scaler.fit_transform(df['Amount'].values.reshape(-1,1))

df.drop(['Amount','Time'], axis = 1, inplace=True)
```

```python
df.head()
```

```python
#getting the scaled amount and time in the front of the data. 
scaled_amount = df['Scaled_Amount']
scaled_time = df['Scaled_Time']

df.drop(['Scaled_Amount','Scaled_Time'], axis=1, inplace=True)
df.insert(0, 'scaled_amount', scaled_amount)
df.insert(1, 'scaled_time', scaled_time)

df.head()
```

```python
# taking X and y for train,test and split

X = df.drop(['Class'], axis = 1)
y = df['Class']
```

```python
#As we have scaled the columns but their is a large imbalance in the data so here we are going to use sklearn 
#undersampaling library, which automatically selects a subset o fthe majority class instances to balance data

#!pip install imbalanced-learn

from imblearn.under_sampling import RandomUnderSampler 

rus = RandomUnderSampler(random_state=42)

X_res, y_res = rus.fit_resample(X,y)
```

```python

X_new = new_df.drop(['Class'], axis = 1)
y_new = new_df['Class']
```

```python
X_new.shape
```

```python
#Removing the unwanted columns using the threshould of 0.8 will remove the higly corr columns. 

correlation_threshould = 0.8

correlation_matrix = X_new.corr()

highly_correlated_cols = set()
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i,j]) > correlation_threshould:
            colname = correlation_matrix.columns[i]
            highly_correlated_cols.add(colname)
            
print('highly correlated columns:', highly_correlated_cols)

new_df_filtered = new_df.drop(columns = highly_correlated_cols)

plt.figure(figsize=(15,8))
sns.heatmap(new_df_filtered.corr(), cmap='coolwarm')
plt.title("Correlation matrixc after removing the highly correlayed columns")
plt.show()
```

Splitting the data for train and test using Hyperparameter. 

```python
#splitting the data for train and test. 
from sklearn.model_selection import train_test_split, GridSearchCV
Features = df1.drop('Class', axis=1)
Target = df1['Class']
```

```python
# now performing train test split for the model 
Features_train, Features_test, Target_train, Target_test = train_test_split(Features, Target, test_size=0.2, random_state=42)
```

```python
#the size of the each train and test 
Features_train.shape, Features_test.shape, Target_train.shape, Target_test.shape
```

# Random Forest 

```python
#Building the model Random Forest classifier
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
```

```python
#Define the parameters for grid search:

param_grid = {
    'n_estimators': [50, 150, 200, 250],
    'max_depth': [None, 5, 10,15],
    'min_samples_split': [2, 4, 8],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2']
}
```

```python
grid_search = GridSearchCV(rf, param_grid, cv=5, scoring='f1', n_jobs=-1)
grid_search.fit(Features_train, Target_train)
```

```python
# Get the best hyperparameters
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)
```

```python
best_rf = RandomForestClassifier(**best_params)
best_rf.fit(Features_train, Target_train)
```

```python
#making predictions on test set 
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
Target_pred = best_rf.predict(Features_test)
accuracy = accuracy_score(Target_test, Target_pred)
confusion = confusion_matrix(Target_test, Target_pred)
print('accuracy:', accuracy)
print('confusion:', confusion)
print("Classifier Report:")
print(classification_report(Target_test, Target_pred))
```

```python
#making predictions on test set 
Target_pred = best_svm.predict(Features_test)
accuracy = accuracy_score(Target_test, Target_pred)
confusion = confusion_matrix(Target_test, Target_pred)
print('accuracy:', accuracy)
print('confusion:', confusion)
print("Classifier Report:")
print(classification_report(Target_test, Target_pred))
```
